```python
def convert_df_column_values_to_frequesncy_to_draw_hist(df, column_name):
  return df[column_name].value_counts().to_dict()
```

```python
import matplotlib.pyplot as plt

# Get the value counts for 'True' and 'False'
ratings_counts = df['Ratings'].value_counts()
true_count = ratings_counts.get(True, 0)  # Get count of True, default to 0 if not found
false_count = ratings_counts.get(False, 0) # Get count of False, default to 0 if not found

#Histogram
plt.bar(['True', 'False'], [true_count, false_count], color="blue", edgecolor="black")
plt.title("Availability of Movies")
plt.xlabel("Is available?")
plt.ylabel("Frequency")
plt.show()
```
